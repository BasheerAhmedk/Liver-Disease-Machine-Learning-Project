```python
# creating log transformation of all features and creating new distplot
Total_Bilirubin = np.log(df1.Total_Bilirubin)
Direct_Bilirubin = np.log(df1.Direct_Bilirubin)
Alkaline_Phosphotase = np.log(df1.Alkaline_Phosphotase)
Alamine_Aminotransferase = np.log(df1.Alamine_Aminotransferase)
Aspartate_Aminotransferase = np.log(df1.Aspartate_Aminotransferase)
Total_Proteins = np.log(df1.Total_Proteins)
Albumin = np.log(df1.Albumin)
Albumin_and_Globulin_Ratio = np.log(df1.Albumin_and_Globulin_Ratio)

plt.figure(figsize=(7.5,4))
sns.distplot(Total_Bilirubin, label='total_bilirubin')
sns.distplot(Direct_Bilirubin, label='direct_bilirubin')
sns.distplot(Alkaline_Phosphotase, label='alkaline_phosphotase')
sns.distplot(Alamine_Aminotransferase, label='alamine_aminotransferase')
sns.distplot(Aspartate_Aminotransferase, label='aspartate_aminotransferase')
sns.distplot(Total_Proteins, label='total_proteins')
sns.distplot(Albumin, label='albumin')
sns.distplot(Albumin_and_Globulin_Ratio, label='albumin_and_globulin_ratio')

plt.title('Log Transformed Features', fontdict={'fontsize': 20})
plt.xlabel('Features')
plt.legend()
plt.show()
```

```python
# min-max scaling by-hand all the features and creating new distplot 
df1['Total_Bilirubin'] = ( Total_Bilirubin - min(Total_Bilirubin) ) / ( max(Total_Bilirubin) - min(Total_Bilirubin) )
df1['Direct_Bilirubin'] = ( Direct_Bilirubin - min(Direct_Bilirubin) ) / ( max(Direct_Bilirubin) - min(Direct_Bilirubin) )
df1['Alkaline_Phosphotase'] = ( Alkaline_Phosphotase - min(Alkaline_Phosphotase) ) / ( max(Alkaline_Phosphotase) - min(Alkaline_Phosphotase) )
df1['Alamine_Aminotransferase'] = ( Alamine_Aminotransferase - min(Alamine_Aminotransferase) ) / ( max(Alamine_Aminotransferase) - min(Alamine_Aminotransferase) )
df1['Aspartate_Aminotransferase'] = ( Aspartate_Aminotransferase - min(Aspartate_Aminotransferase) ) / ( max(Aspartate_Aminotransferase) - min(Aspartate_Aminotransferase) )
df1['Total_Proteins'] = ( Total_Proteins - min(Total_Proteins) ) / ( max(Total_Proteins) - min(Total_Proteins) )
df1['Albumin'] = ( Albumin - min(Albumin) ) / ( max(Albumin) - min(Albumin) )
df1['Albumin_and_Globulin_Ratio'] = ( Albumin_and_Globulin_Ratio - min(Albumin_and_Globulin_Ratio) ) / ( max(Albumin_and_Globulin_Ratio) - min(Albumin_and_Globulin_Ratio) )



plt.figure(figsize=(7.5,4))
sns.distplot(df1.Total_Bilirubin, label='total_bilirubin')
sns.distplot(df1.Direct_Bilirubin, label='direct_bilirubin')
sns.distplot(df1.Alkaline_Phosphotase, label='alkaline_phosphotase')
sns.distplot(df1.Alamine_Aminotransferase, label='alamine_aminotransferase')
sns.distplot(df1.Aspartate_Aminotransferase, label='aspartate_aminotransferase')
sns.distplot(df1.Total_Proteins, label='total_proteins')
sns.distplot(df1.Albumin, label='albumin')
sns.distplot(df1.Albumin_and_Globulin_Ratio, label='albumin_and_globulin_ratio')


plt.title('Scaled Features', fontdict={'fontsize': 20})
plt.xlabel('Features')
plt.legend()
plt.show()
```
